### Load Necessary Libraries

```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

## Load company sales data and store to `sales`

```python
sales=pd.read_csv('company_sales.csv')
sales.head(10)
```

## Use pandas to impute missing valies if there is any

```python
sales.isnull().sum()
```

```python
shirt=sales['Shirt'].mode()[0]
sales['Shirt'].fillna(shirt,inplace=True)
```

```python
sales.head(10)
```

## Exercise 1.Read Bags, Top, Shirt and Pant sales data and show it  using a multiline plot


```python
fig=plt.figure(figsize=(12,5))
plt.plot(sales['month_number'], sales['Bags'], color='blue', label='Bag Sales Data', marker='o')
plt.plot(sales['month_number'], sales['Top'], color='orange', label='Top Sales Data', marker='o')
plt.plot(sales['month_number'], sales['Shirt'], color='green', label='Shirt Sales Data', marker='o')
plt.plot(sales['month_number'], sales['Pant'], color='red', label='Pant Sales Data', marker='o')
plt.xlabel('Month Number')
plt.ylabel('Sales Units in Number')

plt.legend(loc=2)
plt.show
```

## Exercise 2: Read Pant and Shirt sales data of each month and show it using a scatter plot


```python
plt.scatter(sales['month_number'], sales['Shirt'], label='Shirt sales data', color='blue')
plt.scatter(sales['month_number'], sales['Pant'], color='red', label='Pant Sales Data')
plt.legend(loc=2)
plt.xlabel('Month Number')
plt.ylabel('Number of Units Sold')
plt.show()
```

## Exercise 3: Read Jeans and Top product sales data and show it using the bar chart

Note:
-  move x axis of jeans .25 left and Top .25 right and add width
-  plt.bar(sales['month_number']-.25, sales['Jeans'],color='red',width=.30, label="Jeans sales data")

```python
plt.bar(sales['month_number']+.25, sales['Top'], color='blue', width=.30, label='Top Sales Data')
plt.bar(sales['month_number']-.25, sales['Jeans'], color='red', width=.30, label='Jeans Sales Data')
plt.legend(loc=2)
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.title ('Sales Data')
plt.show()
```

## Exercise 4: Read the total profit of each month and show it using the histogram to see the most common profit ranges

```python
plt.hist(sales['total_profit'], label="Profit Data")
plt.xlabel('Profit in Dollars')
plt.title('Profit Data')
plt.legend(loc=1)
plt.show()
```

## Exercise 5: Calculate `total sale` data for last year for each product and show it using a Pie chart

- salesData=[sales['Bags'].sum(),.......]

```python
salesData=[sales['Bags'].sum(), sales['Tie'].sum(), sales['Shirt'].sum(), sales['Jeans'].sum(), sales['Top'].sum(), sales['Pant'].sum()]
```

```python
col=['Bag','Tie','Shirt','Jeans', 'Top', 'Pant']
```

```python
plt.pie(salesData, labels=col, autopct='%.1f%%')
plt.axis('equal')
plt.title('Sales Data')
plt.show()
```

### Exercise 6: Read all product sales data and show it using the stack plot. Save the result as `stack_sales.png`


```python
x= sales['month_number']
y=np.vstack([sales['Bags'], sales['Tie'], sales['Shirt'], sales['Jeans'], sales['Top'], sales['Pant']])
```

```python
labels=['Bag', 'Tie', 'Shirt', "Jeans", 'Top', 'Pant']

stack_sales= plt.stackplot(x,y,labels=labels)
plt.legend(loc='upper left')
plt.show()
```

```python
plt.savefig('stack_sales.png', dpi=300)
```
